```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures , StandardScaler

```

```python
x = np.linspace(0,.5,100)
x.shape
```

```python
ya = np.cos(2*np.pi*x)**2 # y=cos**2(2Ï€x)
ya.shape
```

```python
plt.plot(x,ya)
plt.show()
```

```python
noice = np.random.normal(0,.1,100) # b/w 0 and 0.1 - 100 noice
print(noice)
```

```python
plt.plot(x,ya,color = 'green')
y = ya + noice
#plt.plot(x,y)
plt.scatter(x,y,color = "red")


plt.title("Polynomialdata_add_noice")


# Add a legend
plt.legend()

file_path = r"T:\My Drive\000 ihsan\02 Courses\003GIT\MYProject\Polynomialfeatures\plot\polydataplot.jpg"
plt.savefig(file_path, format='jpg', dpi=300)

plt.show()
```

```python
model = linear_model.LinearRegression()
```

```python
x = x.reshape(-1,1) # error bfore reshape [100,1]
model.fit(x,y)
```

```python
ypredict =model.predict(x)
```

```python
plt.plot(x,ya,color = 'green')
y = ya + noice
plt.plot(x,ypredict)
plt.scatter(x,y,color = "red")


plt.title("Polynomialdata_linear_test")


# Add a legend
plt.legend()

file_path = r"T:\My Drive\000 ihsan\02 Courses\003GIT\MYProject\Polynomialfeatures\plot\polylinear.jpg"
plt.savefig(file_path, format='jpg', dpi=300)

plt.show()
```

```python
plolynoimialfeatures = PolynomialFeatures(degree = 100,include_bias=True) # include_bias = "True" - theeta 0 * x0 - xo = 1 , this feature true ,
#degree = 100 change 2 to 100 plot change correct , degree increase , correct plot probability increase , not accurate in test ,
```

```python
linear = linear_model.LinearRegression(fit_intercept=True)
```

```python
pipeline = Pipeline([("polynomial",plolynoimialfeatures),("linear",linear)]) # call pipe line data go through polynomial and linear
```

```python
pipeline.fit(x,y)
```

```python
y_predict = pipeline.predict(x)
```

```python
y_predict
```

```python
plt.plot(x,ya,color = 'green')
plt.plot(x,y_predict)
plt.scatter(x,y,color = "red")


plt.title("PolynomialFeatures_degree_100")


# Add a legend
plt.legend()

file_path = r"T:\My Drive\000 ihsan\02 Courses\003GIT\MYProject\Polynomialfeatures\plot\polydegree100.jpg"
plt.savefig(file_path, format='jpg', dpi=300)


plt.show()
```

```python
# add scale
scalar = StandardScaler() # more feature scale must
pipelinescale = Pipeline([("polynomial",plolynoimialfeatures),("Scaling",scalar),("linear",linear)]) # call pipe line data go through polynomial and linear
```

```python
pipelinescale.fit(x,y)
```

```python
y_predict = pipelinescale.predict(x)
```

```python
plt.plot(x,ya,color = 'green')
plt.plot(x,y_predict)
plt.scatter(x,y,color = "red")


plt.title("PolynomialFeatures_StandardScaler")


# Add a legend
plt.legend()

file_path = r"T:\My Drive\000 ihsan\02 Courses\003GIT\MYProject\Polynomialfeatures\plot\stdpoly.jpg"
plt.savefig(file_path, format='jpg', dpi=300)

plt.show()
```
